In this notebook, we attempt to classify images of human eyes as male or female. To do this, we train a binary logistic regression model. We work with a data set of ~13,000 RGB images. The array we run the regression on has feature vectors consisting of the pixel values at a position and RGB layer of the image. 

We also run the regression on grayscale 2D arrays. Training and fine tuning the logistic regression model on each and computing the accuracy of the model on the test set gives a prediction accuracy of roughly 80% for the grayscale images and 86% for the RGB images. 

Finally, we run the model on a small number of eyes extracted from headshots of celebrities found on the web. 

```python
#Logistic Regression imports

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import csv
import PIL
import os
import time
import pickle

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils import shuffle
from sklearn.decomposition import PCA
```

```python
#examples of male eyes

#fm_directory = 'female_eyes'
m_directory = 'male_eyes'

for filename in os.listdir(m_directory)[:5]:
    f = os.path.join(m_directory,filename)
    image = PIL.Image.open(f)
    plt.imshow(image)
    plt.show()
```

```python
#read data into list of numpy arrays and resize all the images to a common size

#list of grayscale images
fm_g_L = []
m_g_L = []

#list of full images
fm_L = []
m_L = []

m = 60 #resize images to common size mxm

fm_directory = 'female_eyes'
m_directory = 'male_eyes'

for filename in os.listdir(fm_directory):
    f = os.path.join(fm_directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        image = PIL.Image.open(f)

        image = image.resize((m,m))
            
        fm_L.append(np.array(image).reshape(1,-1))
        fm_g_L.append(np.array(image.convert('L')).reshape(1,-1))
        
for filename in os.listdir(m_directory):
    f = os.path.join(m_directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        image = PIL.Image.open(f)
        
        image = image.resize((m,m))
            
        m_L.append(np.array(image).reshape(1,-1))
        m_g_L.append(np.array(image.convert('L')).reshape(1,-1))
```

```python
#examples of resized and reflected images

for i in range(1):
    a = m_L[i]
    a = a.reshape((m,m,3))
    rev = a[:,::-1,:]
    plt.imshow(a)
    plt.show()
    plt.imshow(rev)
    plt.show()
```

###Logistic Regression

```python
#make labels: female = 0, male  = 1

fm_N = len(fm_L)
m_N = len(m_L)

fm_lab = fm_N*[0]
m_lab = m_N*[1]

print('Female eyes: ' + str(fm_N))
print('Male eyes: ' + str(m_N))
print('Total eyes: ' + str(fm_N + m_N))
```

```python
#randomize and split data into training and test sets

eyes, lab = shuffle(fm_L + m_L, fm_lab + m_lab)
#eyes_g, lab = shuffle(fm_g_L + m_g_L, fm_lab + m_lab)

prop_train = .9
N = int(prop_train*len(eyes))

#augments data by reflecting each image in the training set horizontally
aug = False
if aug:
    aug_train = [eye.reshape((m,m,3))[:,::-1,:].reshape((1,-1)) for eye in eyes[:N]]
    aug_train = np.array(aug_train).reshape((N,3*m**2))
    
    train = np.concatenate((np.array(eyes[:N]).reshape(N,3*m**2), aug_train), axis = 0)
    train_lab = np.array(lab[:N] + lab[:N])
else:
    train = np.array(eyes[:N]).reshape((N,3*m**2))
    train_lab = np.array(lab[:N])


test = np.array(eyes[N:]).reshape((len(eyes) - N,3*m**2))
test_lab = np.array(lab[N:])
```

```python
#correctly reflected images

plt.imshow(eyes[0][0,:].reshape((m,m,3)))
plt.title('Original Eye')
plt.show()

plt.imshow(aug_train[0,:].reshape((m,m,3)))
plt.title('Reflected Eye')
plt.show()
```

```python
#sanity check
print(np.shape(train))
print(np.shape(test))
print(np.shape(train_lab))
print(np.shape(test_lab))
```

```python
#Try PCA before LR
"""
pca = PCA(n_components=500)
pca.fit(train)
train_pca = pca.fit_transform(train)
test_pca = pca.fit_transform(test)
print(np.shape(test_pca))
"""
```

```python
#Train model

#data = train_pca
data = train

model = LogisticRegression(solver='liblinear', random_state=0, max_iter = 100, penalty = 'l2', C = 10**(-7))
model.fit(data,train_lab)

#Best: prop_train = .9, penalty = 'l1', C = .001
```

```python
#Run model on training data

print('Results of Training Data:')
print('-----------------------------')
print('Score: ' + str(model.score(data,train_lab)))
print('-----------------------------')
print('Confusion Matrix:')
print(confusion_matrix(train_lab, model.predict(data))) #false, true, negs, pos
print('-----------------------------')
print(model.predict_proba(data).round(3)[:5])
print('-----------------------------')
print('Prediction:')
model.predict(data)[:5]
```

```python
#Run model on test data
#data_test = test_pca
data_test = test

print('Results of Test Data:')
print('-----------------------------')
print('Score: ' + str(model.score(data_test,test_lab)))
print('-----------------------------')
print('Confusion Matrix:')
print(confusion_matrix(test_lab, model.predict(data_test))) #false, true, negs, pos
print('-----------------------------')
print(model.predict_proba(data_test).round(3)[:5])
print('-----------------------------')
print('Prediction:')
model.predict(data_test)[:5]
```

```python
# save the model to disk
filename = 'model.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(data_test,test_lab)
print(result)
```

```python
#plots of 3 layers of weights corresponding to male eye

plt.imshow(model.coef_[0].reshape((m,m,3))[:,:,0])
plt.title('Red layer')
plt.show()
plt.imshow(model.coef_[0].reshape((m,m,3))[:,:,1])
plt.title('Green layer')
plt.show()
plt.imshow(model.coef_[0].reshape((m,m,3))[:,:,2])
plt.title('Blue layer')
plt.show()

#a = 255*model.coef_[0].reshape((m,m,3))/np.max(model.coef_)
#plt.imshow(PIL.Image.fromarray(,'RGB'))
```

```python
#predictions for first 10 images
for i in range(10):
    if int(model.predict(test)[i]) == 0:
        pred_gen = 'female'
    else:
        pred_gen = 'male'
        
    if int(test_lab[i]) == 0:
        gen = 'female'
    elif int(test_lab[i]) == 1:
        gen = 'male'
        
    print('Eye: ' + gen + '; Predicted: ' + pred_gen + '.')
    a = test[i,:]
    plt.imshow(a.reshape((m,m,3)))
    plt.show()
```

```python
#manually isolate celebrity eyes and run model on them

#coordinates to isolate eyes via array slicing
eye_cd = {'M2':[575,275,75],'M3':[200,400,75],'M5':[125,175,50],'M4':[200,300,50],'M1':[90,95,25],'FM2':[210,150,50],'FM1':[190,220,60]}
celeb_eyes = [] #eyes isolated
celeb_labels = {}
celeb_lab = []


directory = 'Celebs'
for i in range(len(os.listdir('Celebs'))):
    celeb = os.listdir('Celebs')[i]
    f = os.path.join(directory, celeb)
    
    # checking if it is a file
    if os.path.isfile(f):
        
        #generate labes
        name = celeb.split('.')[0]
        if celeb[0] == 'M':
            celeb_labels[name] = 1
            celeb_lab.append(1)
        else:
            celeb_labels[name] = 0
            celeb_lab.append(0)
        
        #extract and store eyes
        image = np.array(PIL.Image.open(f))#.convert('L')

        x,y = eye_cd[name][1], eye_cd[name][0]
        r = eye_cd[name][2]

        eye = PIL.Image.fromarray(image[x:x+r,y:y+2*r])
        celeb_eyes.append(eye.resize((m,m)))
 
for eye in celeb_eyes:
    plt.imshow(eye)
    plt.show()
    
#turn eyes into numpy array to run model on
celeb_eyes_a = np.array([np.array(eye).reshape((1,-1)) for eye in celeb_eyes]).reshape((len(celeb_eyes),-1))
print(np.shape(celeb_eyes_a))
    
#run prediction
loaded_model = pickle.load(open(filename, 'rb'))
print('Score: ' + str(loaded_model.score(celeb_eyes_a,celeb_lab)))
```

```python
#prediction for each eye
for i in range(7):
    if int(model.predict(celeb_eyes_a)[i]) == 0:
        pred_gen = 'female'
    else:
        pred_gen = 'male'
        
    if int(celeb_lab[i]) == 0:
        gen = 'female'
    elif int(celeb_lab[i]) == 1:
        gen = 'male'
        
    print('Eye: ' + gen + '; Predicted: ' + pred_gen + '.')
    a = celeb_eyes_a[i,:]
    plt.imshow(a.reshape((m,m,3)))
    plt.show()
```
